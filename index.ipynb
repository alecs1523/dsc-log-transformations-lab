## Identify Good Candidates for Log Transformation

Below we plot each of the potential numeric features against `SalePrice`:

```python
# Run this cell without changes
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

y = ames["SalePrice"]
X = ames.drop("SalePrice", axis=1)

fig, axes = plt.subplots(nrows=6, ncols=6, figsize=(15,15), sharey=True)

for i, column in enumerate(X.columns):
    # Locate applicable axes
    row = i // 6
    col = i % 6
    ax = axes[row][col]

    # Plot feature vs. y and label axes
    ax.scatter(X[column], y, alpha=0.2)
    ax.set_xlabel(column)
    if col == 0:
        ax.set_ylabel("SalePrice")

fig.tight_layout()
```

### Plot Log Transformed Versions of Features

For each feature that you identified as a good candidate for log transformation, plot the feature vs. `SalePrice` as well as the log transformed feature vs. log transformed `SalePrice`.

```python
# Analyze raw vs log transformed table values for 2ndFlrSF
ames = ames[(ames['2ndFlrSF'] != 0) & (ames['MasVnrArea'] != 0)]

ames_log = ames.copy()
ames_log['log_2ndFlrSF'] = np.log(ames_log['2ndFlrSF'])

ames_log[['2ndFlrSF','log_2ndFlrSF']]
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(7,3))
ax1.hist(ames_log['2ndFlrSF'])
ax2.hist(ames_log['log_2ndFlrSF'],color='orange')
ax1.set_xlabel('2ndFlrSF')
ax2.set_xlabel('Log 2ndFlrSF')
```

```python
# Analyze raw vs log transformed table values for MasVnrArea

ames_log['log_MasVnrArea'] = np.log(ames_log['MasVnrArea'])

ames_log[['MasVnrArea','log_MasVnrArea']]
```

```python
# Plot log transformed and not log transformed graphs
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(7,3))
ax1.hist(ames_log['MasVnrArea'])
ax2.hist(ames_log['log_MasVnrArea'],color='orange')
ax1.set_xlabel('MasVnrArea')
ax2.set_xlabel('Log MasVnrArea')
```

```python
# Analyze raw vs log transformed table values for GarageArea

ames_log['log_GarageArea'] = np.log(ames_log['GarageArea'])

ames_log[['GarageArea','log_GarageArea']]
```

```python
# Plot log transformed and not log transformed graphs
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(7,3))
ax1.hist(ames_log['GarageArea'])
ax2.hist(ames_log['log_GarageArea'],color='orange')
ax1.set_xlabel('GarageArea')
ax2.set_xlabel('Log GarageArea')
```

## Build a Model with Log-Transformed Features and Target

### Data Preparation

Choose up to 3 of the features you investigated, and set up an X dataframe containing the log-transformed versions of these features as well as a y series containing the log-transformed version of the target.

```python
ames_log['log_sales_price'] = np.log(ames_log['SalePrice'])
```

```python
# Your code here - prepare data for modeling
X = ames_log[['2ndFlrSF','GarageArea','MasVnrArea']]
y = ames_log['log_sales_price']
```

### Modeling

Now build a StatsModels OLS model with a log-transformed target as well as log-transformed features.

```python
# Your code here - build a model
model = sm.OLS(y,sm.add_constant(X))
results = model.fit()
print(results.summary())
```
